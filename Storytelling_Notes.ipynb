#### __Plots, Graphs & Charts__

* __Plot__: Points marked on a coordinate system, continuous axes.
* __Graph__: A diagram of a mathematical function or statistical data. Also, nodes connected by edges(ex. network graph)
* __Graphical__: representation of data 

#### _What makes a chart effective?_
1. _Understand_
2. _Create_
3. _Refine_
4. _Present_

##### Things to think about when assessing an effectiveness of a chart :
* Design Execution __⇒__ How well is the chart constructed?
* Contextual Awareness __⇒__  What am I trying to say? To whom? Where?
* High Design Execution __⇒__ High Contextual Awareness ⇒ Good Chart.




####  <u> Types of visual communication  </u>
1. Idea Illustration (Infographics)
2. Everyday Visualization (Reports, Data Journalism)
3. Visual Discovery (Interactive Dashboards, Exploratory Analysis)
4. Idea Generation (Whiteboarding, Brainstorming)




#### <u> Classes of Visualizations </u>
* __Plot__ : points are marked on a coordinate system, continuous axes.
* __Graphs__: a diagram of a mathematical function, but can also be used (loosely) about a diagram of statistical data.
* __Charts__: graphic representation of data, especially when other shapes and symbols (e.g. bars) are used to represent data.
* __Text__: creating imagery through the written word.
* __Speech__: creating imagery through the spoken word.


#### <u>Understand: First Step to an Effective Chart</u> : 
   *  __Understand__ __→__ Create __→__ Refine __→__ Present
##### Recognize a Good Chart :
1. Effective
2. Helps people see the truth, doesn't deceive.
    * Common deception techniques: truncation of y-axis, double y-axis, incorrect use of maps
3. Tells your story


#### <u> Systematically critique charts </u>
1. __Note__ the first few things you see
2. __What is the first idea that forms in your mind?__ Note it and then search for more.
3. __What is the idea being conveyed?__
4. __Does it match the chart's seeming intent?__
5. __Is there anything misleading in the chart?__
6. __Is something missing?__
7. __Note likes, dislikes, wish-I-saws__
8. __Find 3 things you would change__...note them and explain why
9. __Sketch__ and/or protoype your own version and critically evaluate yours.

#### <u>How We Process Visually</u>
1. __Don't assume your audience will read right to left, top to bottom.__ __→__ We don't go in order.

2. __Make what you want the audience to see first stand out!__ __→__ We see first what stands out.

3. __Reduce the number of things on the chart__ __→__ we only see a few things at once.

4. __Gray out the context that needs to remain but not be seen at the same importance as the key idea__ __→__ we only see a few things at once.

5. __Reduce the number of colors__ __→__ we only see a few things at once.

6. __Ensure the meaning is not misrepresented__ __→__ We seek meaning and make connections. If the meaning is not clearly understood, we will make our own.

7. __Use conventions and metaphors__ __→__ Our brains use experience and expectations as cognitive shortcuts so we don't have to consciously process everytime we see something.
    * Red:Green __⇒__ Negative:Postive
    * Red:Blue __⇒__ Hot/Active:Cold/Inactive
    * Hierarchies: Top __⇒__ Bottom
    * Time: Left __⇒__ Right
    * Y-axes: starts at 0
    * Up:Down __⇒__ North:South, Good:Bad
    * Lighter:Darker __⇒__ Emptier:Fuller, Lower:Higher
    * Like Colors __⇒__ Like Items
    * Connected data points __⇒__ relationship from one value to the next, like time, but not categories
    * Categories arranged from Most Extreme __⇒__ Least Extreme
    * Color Gradient/Saturation __⇒__ Progression of Values
    * Distinct Colors __⇒__ Distinct Categories
8. __Limit numbers__ __⇒__  Our eyes are actually really good at math. Most people process values more intuitively via visual representation than numbers or statistics.
