## Create Your Chart: Prepare →  Prototype
 Understand → __Create__  → Refine → Present

1. `Prepare to create`
    * 5 minutes
    * Use NO data
    * Walk away with a sketch of the presentation/visualization(s).
   

2. `Talk and Listen`
    * 15 minutes
    * Talk to a colleague or friend in order for you to explain :
        * Basic information
        * Organize thoughts
        * Provide context
        * Listen to yourself and your friend/colleague for keywords, confusions, etc.

3. `Sketch`
    * 20 minutes
    * Match keywords to approaches
        * (Relationship; Comparison; Distribution; Composition; Maps, Networks, Logic)
    * Answer questions about the types of information.
    * Start sketching: Quick and messy, don't worry about scale, don't add details.

4. `Prototype`
    * 20 Minutes
    * Add in some data
    * Create a non-digital prototype/drawing only as detailed as is needed to take it and convert it to a digital version.

### Brainstorm
`Things to answer/do and document`
1. __Braindump:__ Get a stack of sticky notes and write down potential pieces of content. This is simply a process of getting all of the ideas out of your head and onto paper, without any concern for whether or not they make it into the final presentation.

2. __Refine__ ideas: Group related topics
        * Consider what overall structure will help pull them together in a way that will hopefully make sense to someone else
        * Discard ideas that don't work.

3. __Generate a visual, non-digital outline of the context:__
        * This should demonstrate what ideas/topics will sit where, in what form (text, metric, chart, ...)
        
`Notes about brainstorming`

* Do NOT skip this step and go straight to the tools.
        * This step allows up to physically look at each potential piece of content and ask ourselves: does this help us get our main point across?

* __Sticky Notes__ : 
       
       
 * Their size requires you to be concise, they are easy to arrange and re-arrange to explore different flows.
        
 * They require minimal time investment to capture a thought leading to less time wasted on an idea you may not use
 * They lead to shorter and more effective communications.
 * Each sticky note could eventually become a slide in a deck, a section in a report, or possibly a graph on a dashboard

### Sketch

` Things to answer/do and document`
1. What type of information are you communicating?
2. Map keywords to approaches/types of visualization to identify the chart type(s) that best convey your message.
3. Sketch chart(s) quickly, without scale, without details. 

* `Time: 20 minutes `
* `Type of information:`
    * Quantitative or Qualitative data
    * How is it exposed: numerically, textually, geospatially, relationally?
    * Can the data be measured or counted (discrete data (counted), continuous data (measured))?
    * Does order matter (ordered data, unordered data)?
    * Are you visualizing observations or statistics about the observations (observations, statistics)?
* `Keywords to approaches:` <a href="https://public.tableau.com/profile/andy.kriebel#!/vizhome/VisualVocabulary/VisualVocabulary">Andy Kriebel's Visual Vocabulary</a>
    * __Relationship:__ correlation, relationship, affect, change with, related
    * __Comparison:__ before/after, categories, compare, contrast, over time, peaks, rank, trend, types, valleys
    * __Distribution:__ cluster, distributed, from/to, plotted, points, spread, relative to, transfer
    * __Composition:__ components, group, makes up, of the whole, parts, percentage, pieces, portion, proportion, slices, total
    * __Maps, Networks, Logic:__ cluster, complex, connections, group, hierarchy, if/then, network, relationships, places, routes, structure, paths, organize, space, yes/no
