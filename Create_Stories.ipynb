## Creating a Story

A Tableau Story is a sequence of visualizations that work together to convey information. You can create stories to tell a data narrative, provide context, demonstrate how decisions relate to outcomes, or to simply make a compelling case. ~ <a href= "https://help.tableau.com/current/pro/desktop/en-us/stories.htm">Stories Tableau Help</a>

* A story is a separate sheet that contains separate sheets within it.
    * These are conveyed through `story points` which are shown as buttons in the center at the top of each page throughout the story.
    * The user interacts by clicking on the story points.

### <a href="https://help.tableau.com/current/pro/desktop/en-us/story_workspace.htm">Story Elements</a>
* `Options for adding a new story point:`
    * Choose Blank to add a new point
    * Duplicate to use the current story point as the starting place for your next point.
* `The Story pane:` 
    * Use this pane to drag dashboards, sheets, and text descriptions to your story sheet.
    * Also where you set the size of your story and display or hide the title.
* `The layout pane:` 
    * Where you choose your navigator style and display or hide the forward and back arrows.
* `The Story menu:`
    * Format the story or copy or export the current story point as an image.
    * Can also clear the entire story here or show or hide the navigator and story title.
* `The Story toolbar:` 
    * appears when you mouse-over the navigator area.
    * Use it to:
        * Revert changes
        * Apply updates to a story point
        * Delete a story point
        * Create a new story point out of the current, customized one.
        
* `The navigator:` 
    * Edit and organize your story points.
    * How your audience will step through your story
    * To change the style of the navigator, use the Layout pane.
